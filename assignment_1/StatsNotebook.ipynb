```python
import matplotlib.pyplot as plt
```

```python
eth1 = []
eth0 = []
with open('NETWORK_STAT.txt', 'r') as ns:
    for line in ns:
        if 'eth1' in line: eth1.append(line.strip())
        if 'eth0' in line: eth0.append(line.strip())
```

```python
counts1 = [line.split() for line in eth1]
counts0 = [line.split() for line in eth0]
```

```python
vals1 = [(x[0], float(x[3]) + float(x[4])) for x in counts1]
vals0 = [(x[0], float(x[3]) + float(x[4])) for x in counts0]
```

```python
vals
```

```python
plt.plot([x[1] for x in vals1], label='eth1')
plt.plot([x[1] for x in vals0], label='eth0')
plt.title('pkts/s on network interfaces during job execution')
plt.xlabel('time elapsed (seconds)')
plt.ylabel('pkts/s')
plt.legend()
```

```python
all_cpus = []
with open('MPSTAT.txt', 'r') as ns:
    for line in ns:
        if 'all' in line: all_cpus.append(line.strip())
```

```python
plt.plot([float(x.split()[3]) for x in all_cpus])
plt.title('worker cpu utilisation')
plt.xlabel('time elapsed')
plt.ylabel('%')
```
